```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np
import astropy
from astropy import units as u
from astropy.coordinates import SkyCoord


fig = plt.figure()
subplot = fig.add_subplot(projection='3d')

scale = 1
subplot.set_zlim(-scale, scale)
subplot.set_ylim(-scale, scale)
subplot.set_xlim(-scale, scale)

# Plot origin. Sun / Earth
subplot.scatter(0, 0, 0, c='red')
subplot.text(0, 0, 0, 'Earth')

# M45 03h 47m 24s = 56.85Â°    +24Â° 07â€² 00â€³    65.8833
rho = .444     # Distance in kly
theta = 0.99221968
phi = 1.149880507
c = SkyCoord('03h47m24s', '+24d07m00s')
print(c.dec.degree)
print(c.ra.degree)
print(c.dec.radian)
print(c.ra.radian)

x1, y1, z1 = astropy.coordinates.spherical_to_cartesian(rho, c.dec.radian, c.ra.radian)
print(x1, y1, z1)

x = rho * np.sin(phi) * np.cos(theta)
y = rho * np.sin(phi) * np.sin(theta)
z = rho * np.cos(phi)
print(x,y,z)
subplot.scatter(x, y, z)
subplot.text(x, y, z, 'M45')

# M40 
#rho = .510     # Distance in kly
#theta = 0.99221968
#phi = 1.149880507
#x = rho * np.sin(phi) * np.cos(theta)
#y = rho * np.sin(phi) * np.sin(theta)
#z = rho * np.cos(phi)
#subplot.scatter(x, y, z)
#subplot.text(x, y, z, 'M40')


plt.show()
```
